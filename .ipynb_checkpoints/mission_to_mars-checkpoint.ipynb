```python
# Import Splinter, BeautifulSoup, and Pandas
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up driver
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

# NASA Mars News

```python
# Visit the Mars
url = 'https://redplanetscience.com/'
browser.visit(url)

time.sleep(1)

# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

# Find Website Headlines 

```python
# Convert the browser html to a soup object 
html = browser.html
news_soup = bs(html, 'html.parser')
print(news_soup.prettify())
```

```python
# Scrape the Mars first article and description
news_title = news_soup.select("#news .content_title")[0].text
news_description = news_soup.select("#news .article_teaser_body")[0].text

print(news_title)
print(news_description)
```

```python
# hemisphere
url = 'https://marshemispheres.com/'
browser.visit(url)
```

```python
# Create a list to hold the images and titles
hemisphere_image_urls = []

# get a list of all the hemispheres
links = browser.find_by_css('a.product-item img')

# loop through those links, click the link, find the sample anchor, return the href
for i in range(len(links)):
    #hemisphere information dictionary
    hemisphereInfo = {}
    
    # we have to find the elements on each loop to avoid a stale element exception
    browser.find_by_css('a.product-item img')[i].click()
    time.sleep(2)
    
    # we find the sample image anchor tag and extract the href
    sample = browser.links.find_by_text('Sample').first
    hemisphereInfo["img_url"]= sample['href']
    
    # get the hemisphere title
    hemisphereInfo['title'] = browser.find_by_css('h2.title').text
    
    # append hemisphere object to list
    hemisphere_image_urls.append(hemisphereInfo)
    
    # finally, we navigate backwards
    browser.back()
```
